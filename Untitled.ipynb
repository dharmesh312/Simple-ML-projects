```python
    traindf=pd.read_csv('my.tsv',sep='\t')
#     traindf.describe()
#     print (traindf.loc[traindf['essay_set']==8].head())
    traindf=traindf.iloc[:,[0,1,2,6,9]]
    print (traindf.loc[traindf['essay_set']==8].index)
```

```python
count=0
    ##################### feature extraction#######################
traindf['word_tokenize'] =traindf.apply(lambda row: len(word_tokenize(row['essay'])), axis=1)
traindf['sent_tokenize'] =traindf.apply(lambda row: len(sent_tokenize(row['essay'])), axis=1)

nounsArr=[]
advArr=[]
adjArr=[]
similar=[]
for i in range(7,8):
    print ('genration feature for set id=' ,i )
    traindfTemp=traindf.loc[traindf['essay_set']==i]
    bestEssay=traindfTemp.loc[traindfTemp['domain1_score']==max(traindfTemp['domain1_score'])]['essay']
    nouns=set()
    nouns2=set()
#   getting the nouns of the best essay the one which has the highest scores  
    for essays in bestEssay:
        sentences = sent_tokenize(essays) #tokenize sentences
        for sentence in sentences:
            for word,pos in nltk.pos_tag(nltk.word_tokenize(str(sentence))):
                if ((pos == 'NN' or pos == 'NNP' or pos == 'NNS' or pos == 'NNPS') and  word.isalpha() and word!='@')  :
                    nouns.add(word.lower())

    
    syno=set()
    for words in list(nouns):
        for syn in wordnet.synsets(words):
            for l in syn.lemmas():
                syno.add(l.name())


    pickle.dump(syno,open('syno'+str(i),'wb'))
    for x in traindfTemp.index:
        sentences = sent_tokenize(traindfTemp.loc[x]['essay']) #tokenize sentences
        tempNouns=set()
        syno2=set()
        for sentence in sentences:
            for word,pos in nltk.pos_tag(nltk.word_tokenize(str(sentence))):
                if ((pos == 'NN' or pos == 'NNP' or pos == 'NNS' or pos == 'NNPS') and  word.isalpha() and word!='@')  :
                    tempNouns.add(word)


        for words in list(tempNouns):
            for syn in wordnet.synsets(words):
                for l in syn.lemmas():
                    syno2.add(l.name())


        score=0
        for x in list(syno2):
            if x in syno:
                score=score+1
#         print (len(similar),score)
        similar.append(score)

    print ('genration feature for set id=' ,i )
    for x in traindfTemp.index:
        nouns=0
        adv=0
        adj=0
        sentences = sent_tokenize(traindfTemp.loc[x]['essay']) #tokenize sentences
        for sentence in sentences:
            for word,pos in nltk.pos_tag(nltk.word_tokenize(str(sentence))):
                if ((pos == 'NN' or pos == 'NNP' or pos == 'NNS' or pos == 'NNPS') and  word.isalpha() and word!='@')  :
                    nouns=nouns+1
                elif ((pos == 'RB' or pos == 'RBR' or pos == 'RBS') and  word.isalpha() and word!='@')  :
                    adv=adv+1
                elif ((pos == 'JJ' or pos == 'JJR' or pos == 'JJSS' ) and  word.isalpha() and word!='@')  :
                    adj=adj+1

        nounsArr.append(nouns)
        advArr.append(adv)
        adjArr.append(adj)
    traindf=traindfTemp
        
traindf['nouns']=nounsArr
traindf['adverb']=advArr
traindf['adjective']=adjArr
traindf['similarity']=similar

pickle.dump(traindf,open('traindf','wb'))
```
